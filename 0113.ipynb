```python
# get yahoo finance news using requests.get
response = requests.get('https://finance.yahoo.com/news/')
response.status_code # for test

response.text # for test

# Start HTML Parsing with BeautifulSoup
soup = BeautifulSoup(response.text, 'lxml')
news_ul = soup.find('div', attrs={'id':"Fin-Stream"}).ul

# find and append all <li>
lis = news_ul.find_all('li')
len(lis) # for test

# define process using lis[0]
news_title = lis[0].h3.text
news_abstract = lis[0].p.text

# Do things to all li
news = {}
for li in lis:
   news_title = li.h3.text
   news_abstract = li.p.text
   news.setdefault('articles', []).append({
       'title': news_title,
       'abstract': news_abstract,
   })
news
```

```python
# request rottentomatoes main page
response = requests.get('https://www.rottentomatoes.com/')
response.status_code # for test
response.text # for test

# Start parsing with BeautifulSoup
soup = BeautifulSoup(response.text, 'lxml')
# find streaming chart <ul> using find_all
uls = soup.find_all('ul', attrs={'slot':'list-items'})
# select first one of uls
streaming_chart = uls[0]
# find all li in streaming_chart
lis = streaming_chart.find_all('li')
len(lis) # for check

# Create empty list
result = []
# for all li in lis
for li in lis:
  # find and store data
  uri = 'https://www.rottentomatos.com' + li.find('a')['href']
  spans = li.find_all('span')
  movie_name = spans[0].text
  movie_score = spans[2].text.strip('\n %')
  #print(uri, movie_name, movie_score) # for test
  result.append({
      'uri': uri,
      'movie_name': movie_name,
      'movie_score': movie_score,
      })
```
